```python
# initializing running_total and count
running_total = 0
count = 0
# iterate through DataFrame rows
for index, row in t1.iterrows():   # iterrows() generates as an iterator object that yields index and row data as a series
    if np.isnan(row['Cost']):      # check Cost is nan
        if count > 0:              # make sure we are not dividing by zero
            running_average = running_total/count   # replace the missing value with running average
        else:
            running_average = 0
        t1.at[index,'Cost'] = running_average
    else:
        running_total += row['Cost']
        count += 1
# Set display option to show float numbers in decimal format
pd.set_option('display.float_format', '{:.2f}'.format)
t1
```

```python
#split the name column into first and last name
t2[['First Name','Last Name']]= t2['Name'].str.split(expand = True)
#drop the original name column
t2 = t2.drop(columns = ['Name'])
t2
```

```python
Final = t1.merge(t2, on ='ID', how = 'outer').merge(t3, on = 'ID', how = 'outer')
Final
```

```python
#Define a function to calculate bonus
def calculate_bonus(row):
    if row['Status'] == 'Finished':     # check the status of the project
        return row['Cost']*0.05         # 5% bonus
    else:
        return 0
Final['Bonus'] = Final.apply(calculate_bonus, axis = 1)  # axis = 1=> function receives each row as a series(function should be applied to each row)
Final
```

```python
#defining demote function
def demote(row):
    if row['Status'] == 'Failed':            #checking status
        return row['Designation Level']-1
    else:
        return row['Designation Level']
Final['Designation Level'] = Final.apply(demote, axis = 1)       #appling function to DataFrame
Final = Final[Final['Designation Level'] <= 4]                   #deleting employee record whose Designation level is above 4
Final
```

```python
# defining salutations function
def salutations(row):
    if row['Gender'] == 'M':
        return 'Mr.' +row['First Name']
    else:
        return 'Mrs.'+row['First Name']
# apply the salutations function to first name column
Final['First Name'] = Final.apply(salutations, axis = 1)
# drop gender column
Final = Final.drop(columns = ['Gender'])
Final
```

```python
# promote designation level based on age
def promote(row):
    if row['Age'] > 29:
        return row['Designation Level'] + 1
    else:
        return row['Designation Level']
Final['Designation Level'] = Final.apply(promote, axis =1)
Final
```

```python
# adding cost of all projects for each employee
TotalProjCost = Final.groupby(['ID','First Name'],as_index = False)['Cost'].sum()
TotalProjCost.rename(columns ={'Cost':'Total cost'}, inplace = True)
TotalProjCost
```
